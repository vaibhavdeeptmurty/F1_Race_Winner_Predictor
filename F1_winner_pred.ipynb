```python
from xgboost import XGBClassifier
```

```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
```

```python
X = train.drop(columns=['race_winner'])
y = train['race_winner']
```

```python
model = XGBClassifier(
    n_estimators=200,
    max_depth=5,
    learning_rate=0.04,
    random_state=42
)

model.fit(X, y)
```

```python
test = pd.read_csv('test.csv')
```

```python
X_test = test.drop(columns=['race_winner'])
y_test = test['race_winner']
```

```python
win_probs = model.predict_proba(X_test)[:, 1]

win_probs = win_probs / win_probs.sum()
win_probs = win_probs * 100

print(win_probs)
```

```python
driver_map={"George Russell":63,"Kimi Antonelli":12,"Charles Leclerc":16,"Lewis Hamilton":44,"Lando Norris":1,"Oscar Piastri":81,"Max Verstappen":3,"Isack Hadjar":6}
```

```python
team_map={"Mercedes":1,"Ferrari":2,"McLaren":3,"RedBull Racing":4}
```

```python
reverse_driver_map = {v: k for k, v in driver_map.items()}
reverse_team_map = {v: k for k, v in team_map.items()}
```

```python
test["driver"]=test["driver"].map(reverse_driver_map)
test["team"]=test["team"].map(reverse_team_map)
```

```python
test["win_probability"]=win_probs
```

```python
test.head(10)
```

```python
predictions = (
    test[["driver", "team", "win_probability"]]
    .sort_values("win_probability", ascending=False)
    .reset_index(drop=True)
)
```

```python
print("\n🏁 Austrian GP Winner Prediction\n")
for pos, row in predictions.iterrows():
    print(
        f"P{pos+1:<2} | "
        f"{row['driver']:<20} | "
        f"{row['team']:<20} | "
        f"{row['win_probability']:.3f}%"
    )
```
